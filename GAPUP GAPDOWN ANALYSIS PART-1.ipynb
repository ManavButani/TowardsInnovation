```python
import pandas as pd
```

```python
df = pd.read_csv('./nifty_ohh/NIFTY 50 - Minute data 2015 to Aug 2024.csv')
df.index = pd.to_datetime(df['date'])
```

```python
# df.index.name = df.index.name.lower()
df.columns = df.columns.str.lower()

df['year'] = df.index.year.astype(int)
df['month'] = df.index.month.astype(int)
df['day'] = df.index.day.astype(int)
df['hour'] = df.index.hour.astype(int)
df['minute'] = df.index.minute.astype(int)
df['day_name'] = df.index.day_name()
df['max'] = df[['open', 'high', 'low', 'close']].max(axis=1)
df['min'] = df[['open', 'high', 'low', 'close']].min(axis=1)
df['difference'] = df['max'] - df['min']
df['candle_body'] = df['open'] - df['close']
df['high_shadow'] = df['high'] - df[['open', 'close']].max(axis=1)
df['low_shadow'] = df['low'] - df[['open', 'close']].min(axis=1)
df['low-open'] = df['low'] - df['open']
df['high-open'] = df['high'] - df['open']
df['trend'] = df['candle_body'].apply(lambda x: "DOWN" if x < 0 else "ZERO" if x == 0 else "UP")
df = df.sort_values(by=['year', 'month', 'day'], ascending=[True, True, True])
```

```python
dfs = [group for _, group in df.groupby(['year','month', 'day'])]
```

```python
df_1d = pd.DataFrame()

for df in dfs:
    ohlc_data = pd.DataFrame({
    'high' : [df['max'].max()],
    'low' : [df['min'].min()],
    'open' : [df.iloc[0].open],
    'close' : [df.iloc[-1].close],
    'year': [df.iloc[0].year],
    'month': [df.iloc[0].month],
    'day': [df.iloc[0].day],
    })

    ohlc_data.index = pd.to_datetime([f"{int(df.iloc[-1].year)}-{int(df.iloc[-1].month)}-{int(df.iloc[-1].day)}"])
    df_1d = pd.concat([df_1d, ohlc_data])
```

```python
# Create a new column with the difference between high of the current row and open of the next row
df_1d['close_open_diff'] = df_1d['close'] - df_1d['open'].shift(-1)
```

```python
df_1d = df_1d[~df_1d['year'].isin([2019,2020,2021])]
```

```python
max_value = df_1d['close_open_diff'].max()
min_value = df_1d['close_open_diff'].min()
mean_value = df_1d['close_open_diff'].mean()
median_value = df_1d['close_open_diff'].median()
mode_value = df_1d['close_open_diff'].mode()

print(f"Max: {max_value}")
print(f"Min: {min_value}")
print(f"Mean: {mean_value}")
print(f"Median: {median_value}")
print(f"Mode: {mode_value}")
```

```python
import matplotlib.pyplot as plt

# Replace 'column_name' with your numeric column
plt.figure(figsize=(10, 6))
plt.hist(df_1d['close_open_diff'], bins=40, edgecolor='k', alpha=0.7)
plt.title('Histogram of close_open_diff')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
# Define a function to remove outliers based on IQR method
def remove_outliers_iqr(manav, column):
    Q1 = manav[column].quantile(0.25)
    Q3 = manav[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Filter out rows where the column value is within the bounds
    filtered_manav = manav[(manav[column] >= lower_bound) & (manav[column] <= upper_bound)]
    return filtered_manav

# Remove outliers from 'high_open_difference' column
df_filtered = remove_outliers_iqr(df_1d, 'close_open_diff')

# Plot the distribution of the filtered data
df_filtered['close_open_diff'].plot(kind='hist', bins=30, title="Distribution Without Outliers")
```

```python
# Filter the DataFrame to exclude values between -10 and 10
df_filtered = df_filtered[(df_filtered['close_open_diff'] < -10) | (df_filtered['close_open_diff'] > 10)]
```
