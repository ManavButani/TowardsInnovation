```python
import yfinance as yf
import pandas as pd
```

```python
ticker_symbol = 'BTC-USD'
df = yf.download(ticker_symbol, start='2024-01-01', end='2024-10-01', interval='60m')
df.index.name = df.index.name.lower()
df.columns = df.columns.str.lower()
df['year'] = df.index.year
df['month'] = df.index.month
df['hour'] = df.index.hour
df['day'] = df.index.day
df['day_name'] = df.index.day_name()
```

```python
tuesday_data = df[df['day_name']=='Tuesday'].sort_values(by=['day','volume'], ascending=[True,False])
```

```python
tuesday_max_volume = max(tuesday_data['volume'])
```

```python
january_tuesday = tuesday_data[tuesday_data['month']==1]
february_tuesday = tuesday_data[tuesday_data['month']==2]
march_tuesday = tuesday_data[tuesday_data['month']==3]
april_tuesday = tuesday_data[tuesday_data['month']==4]
may_tuesday = tuesday_data[tuesday_data['month']==5]
june_tuesday = tuesday_data[tuesday_data['month']==6]
july_tuesday = tuesday_data[tuesday_data['month']==7]
aug_tuesday = tuesday_data[tuesday_data['month']==8]
sep_tuesday = tuesday_data[tuesday_data['month']==9]
oct_tuesday = tuesday_data[tuesday_data['month']==10]
```

```python
january_tuesday
```

```python
january_dfs = [group.assign(volume=group['volume'] / tuesday_max_volume) for _, group in january_tuesday[['day','hour','volume']].groupby('day')]

total_days = len(january_dfs)

most_repeating_january, least_repeating_january = [], []
for january_df in january_dfs:
    print("~~~~~~~~~~~~~~~~",january_df['day'].iloc[0]," January ~~~~~~~~~~~~~~~~~~")
    result = {hour: volume for hour, volume in zip(january_df['hour'], january_df['volume'])}
    print(result)
    print("==========")
    hours = list(result.keys())
    print(hours)
    most_repeating_january.extend(hours[:len(hours)//2])
    least_repeating_january.extend(hours[len(hours)//2:])

from collections import Counter
most_repeating_january_counts = Counter(most_repeating_january)
least_repeating_january_counts = Counter(least_repeating_january)

print("*"*20)
print(most_repeating_january_counts)
print(least_repeating_january_counts)
print("*"*20)

make_trades = [key for key, value in most_repeating_january_counts.items() if value >= total_days]
avoid_trades = [key for key, value in least_repeating_january_counts.items() if value >= total_days]

print("Make Trades",make_trades)
print('Avoid Trades',avoid_trades)
```
