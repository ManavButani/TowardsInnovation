```python
import pandas as pd
import mplfinance as mpf
```

```python
df = pd.read_csv('NIFTY 50 - 1 Day data 2015 to Aug 2024 GO CORONA.csv')
df_5 = pd.read_csv('NIFTY 50 - 1m data 2015 to Aug 2024 first 5 minute single candle.csv')
```

```python
df_2024 = df[(df['year']==2024)]
df_5_2024 = df_5[(df_5['year']==2024)]
df_2024['source']='B'
df_5_2024['source']='A'
```

```python
df.head(15)
```

```python
df_5_2024.head(10)
```

```python
merged_df = pd.concat([df_5_2024, df_2024])

merged_df = merged_df.sort_values(by=['month','day', 'source'], ascending=[True,True, True])

merged_df = merged_df.iloc[5:-5,:]
```

```python
merged_df.head(10)
```

```python
import os
from datetime import datetime
```

```python
# df_60_min = pd.DataFrame()

for i in range(0, len(merged_df)-1, 6):
    df_chunk = merged_df.iloc[i:i+6] # Get 2 rows in each iteration
    directory_path = f"{os.getcwd()}/gap_vs_first_five_candle/{str(df_chunk['year'].iloc[0])}_{str(df_chunk['month'].iloc[0])}"
    if not os.path.exists(directory_path):
        os.makedirs(directory_path)
    ohlc = df_chunk.iloc[1:,:]
    ohlc.index = pd.to_datetime([f"{int(df.iloc[-1].year)}-{int(df.iloc[-1].month)}-{int(df.iloc[-1].day)}"]*5)
    mpf.plot(ohlc, type='candle', style='charles', title=f'{df_chunk["close_open_diff"].iloc[0]}', ylabel='Price', savefig=os.path.join(directory_path,str(df_chunk['day'].iloc[-1])))
```
