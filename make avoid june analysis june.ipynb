```python
import pandas as pd
```

```python
df = pd.read_csv('trend_2024.csv')
```

```python
july = df[df['month']==6]
```

```python
median_volume_of_july = july['volume'].median()
```

```python
july_dfs = [group for _, group in july[['day','hour','volume', 'trend']].groupby('day')]
total_days = len(july_dfs)

most_repeating_july, least_repeating_july = [], []
high_volume_hours, low_volume_hours = [], []
for july_df in july_dfs:
    july_up_count = 0
    july_down_count = 0
    july_up_volume = 0
    july_down_volume = 0
    july_up_hours = []
    sorted_hours = []
    july_sorted_up_hours = []
    july_down_hours = []
    july_sorted_down_hours = []
    high_volume_hours_, low_volume_hours_ = [], []

    july_volume_median = july_df['volume'].median()

    print("~~~~~~~~~~~~~~~~",july_df['day'].iloc[0]," july ~~~~~~~~~~~~~~~~~~")
    july_hour_volume = {hour: volume for hour, volume in zip(july_df['hour'], july_df['volume'])}
    july_hour_trend = {hour: trend for hour, trend in zip(july_df['hour'], july_df['trend'])}

    print(july_hour_trend)
    print(july_hour_volume)
    print("-----------------")


    for hour, trend in july_hour_trend.items():
        volume_ = july_hour_volume.get(hour)
        if volume_ < july_volume_median:
            low_volume_hours.append(hour)
        else:
            high_volume_hours.append(hour)

        if trend == "UP":
            july_up_count+=1
            july_up_volume+=volume_
            july_up_hours.append(hour)
        else:
            july_down_count+=1
            july_down_volume+=volume_
            july_down_hours.append(hour)

    for hour in dict(sorted(july_hour_volume.items(), key=lambda item: item[1])).keys():
        sorted_hours.append(hour)
        if july_hour_trend.get(hour) == "UP":
            july_sorted_up_hours.append(hour)
        else:
            july_sorted_down_hours.append(hour)

    print("UP:::", july_up_count, july_up_volume)
    print("DOWN:::", july_down_count, july_down_volume)
    print("-----------------")

    print("UP Hours:::", july_up_hours)
    print("DOWN Hours:::", july_down_hours)
    print("-----------------")

    print("SORTED UP Hours:::", july_sorted_up_hours)
    print("SORTED DOWN Hours:::", july_sorted_down_hours)
    print("-----------------")
    print("SORTED HOURS ASCENDING:::", sorted_hours)
    print("SORTED HOURS DESCENDING:::", list(reversed(sorted_hours)))

    print("-----------------")
    high_volume_hours.extend(high_volume_hours_)
    low_volume_hours.extend(low_volume_hours_)
    print("Make Trades", high_volume_hours_)
    print("Avoid Trades", low_volume_hours_)
```

```python
from collections import Counter
most_repeating_counts = Counter(high_volume_hours)
least_repeating_counts = Counter(low_volume_hours)
```

```python
most_repeating = dict(sorted(most_repeating_counts.items(), key=lambda item: item[0]))
least_repeating = dict(sorted(least_repeating_counts.items(), key=lambda item: item[0]))
```
