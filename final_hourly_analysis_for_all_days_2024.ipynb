## Suitable Times for trading

# Sunday to Saturday is belows as per volume


```python
ticker_symbol = 'BTC-USD'
df = yf.download(ticker_symbol, start='2024-01-01', end='2024-10-07', interval='60m')
df.index.name = df.index.name.lower()
df.columns = df.columns.str.lower()
df['year'] = df.index.year
df['month'] = df.index.month
df['hour'] = df.index.hour
df['day'] = df.index.day
df['day_name'] = df.index.day_name()
```

```python
new_df = df.groupby(['day_name', 'hour']).agg({'volume': 'sum'}).reset_index().sort_values(by='volume', ascending=False)
```

```python
sunday_data = new_df[new_df['day_name']=='Sunday']
monday_data = new_df[new_df['day_name']=='Monday']
tuesday_data = new_df[new_df['day_name']=='Tuesday']
wednesday_data = new_df[new_df['day_name']=='Wednesday']
thursday_data = new_df[new_df['day_name']=='Thursday']
friday_data = new_df[new_df['day_name']=='Friday']
saturday_data = new_df[new_df['day_name']=='Saturday']
```

```python
sunday_hours_volume_sum = sunday_data.groupby(['hour']).agg({'volume': 'sum'}).sort_values(by='volume', ascending=False) 
monday_hours_volume_sum = monday_data.groupby(['hour']).agg({'volume': 'sum'}).sort_values(by='volume', ascending=False) 
tuesday_hours_volume_sum = tuesday_data.groupby(['hour']).agg({'volume': 'sum'}).sort_values(by='volume', ascending=False) 
wednesday_hours_volume_sum = wednesday_data.groupby(['hour']).agg({'volume': 'sum'}).sort_values(by='volume', ascending=False) 
thursday_hours_volume_sum = thursday_data.groupby(['hour']).agg({'volume': 'sum'}).sort_values(by='volume', ascending=False) 
friday_hours_volume_sum = friday_data.groupby(['hour']).agg({'volume': 'sum'}).sort_values(by='volume', ascending=False) 
saturday_hours_volume_sum = saturday_data.groupby(['hour']).agg({'volume': 'sum'}).sort_values(by='volume', ascending=False) 
```
