```python
btc_data_from_2019_to_2024_part_1_1_d['year'] = btc_data_from_2019_to_2024_part_1_1_d.index.year
btc_data_from_2019_to_2024_part_1_1_d['month'] = btc_data_from_2019_to_2024_part_1_1_d.index.month
btc_data_from_2019_to_2024_part_1_1_d['day_name'] = btc_data_from_2019_to_2024_part_1_1_d.index.day_name()
```

```python
btc_data_from_2019_to_2024_part_1_1_d
```

```python
btc_data_from_2019_to_2024_part_1_1_d['max'] = btc_data_from_2019_to_2024_part_1_1_d[['open', 'high', 'low', 'close']].max(axis=1)
btc_data_from_2019_to_2024_part_1_1_d['min'] = btc_data_from_2019_to_2024_part_1_1_d[['open', 'high', 'low', 'close']].min(axis=1)
btc_data_from_2019_to_2024_part_1_1_d['difference'] = btc_data_from_2019_to_2024_part_1_1_d['max'] - btc_data_from_2019_to_2024_part_1_1_d['min']
```

```python
btc_data_from_2019_to_2024_part_1_1_d
```

```python
year_volume_chart = btc_data_from_2019_to_2024_part_1_1_d.groupby(btc_data_from_2019_to_2024_part_1_1_d.year).agg({'volume': 'sum'})
```

```python
year_volume_chart
```

```python
month_volume_chart = btc_data_from_2019_to_2024_part_1_1_d.groupby(btc_data_from_2019_to_2024_part_1_1_d.month).agg({'volume': 'sum'})
```

```python
month_volume_chart
```

```python
month_volume_chart = month_volume_chart.sort_values(by='volume', ascending=True)
```

```python
# Group by the date index and sum the volumes
# df_daily_volume = btc_data_from_2019_to_2024_part_1_1_d.groupby(btc_data_from_2019_to_2024_part_1_1_d.index.date)['volume'].sum()
```

```python
month_year_volume_analysis = btc_data_from_2019_to_2024_part_1_1_d.groupby(['month', 'year']).agg({'volume': 'sum'}).reset_index()
```

```python
month_year_volume_analysis
```

```python
month_year_volume_analysis = month_year_volume_analysis.sort_values(by='volume', ascending=True)
```

```python
month_year_volume_analysis
```

```python
# Ignoring period of corona

without_corona_df = btc_data_from_2019_to_2024_part_1_1_d[~btc_data_from_2019_to_2024_part_1_1_d['year'].isin([2020, 2021])]
```

```python
month_year_volume_analysis = without_corona_df.groupby(['month', 'year']).agg({'volume': 'sum'}).reset_index()
```

```python
# month_year_volume_analysis = month_year_volume_analysis.sort_values(by='volume', ascending=True)
```

```python
month_year_volume_analysis
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a line plot
plt.figure(figsize=(10, 6))
sns.lineplot(data=month_year_volume_analysis, 
             x=month_year_volume_analysis['year'].astype(str) + '-' + month_year_volume_analysis['month'].astype(str), 
             y='volume', 
             marker='o')

plt.title('Monthly Volume Analysis')
plt.xlabel('Year-Month')
plt.ylabel('Total Volume')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()

plt.show()
```

```python
# Create a combined 'Year-Month' column for better x-axis labels
month_year_volume_analysis['year_month'] = month_year_volume_analysis['year'].astype(str) + '-' + month_year_volume_analysis['month'].astype(str)

# Line Plot
sns.lineplot(data=month_year_volume_analysis, x='year_month', y='volume', marker='o')
plt.title('Monthly Volume Analysis')
plt.xlabel('Year-Month')
plt.ylabel('Total Volume')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()

# Bar Plot
plt.figure(figsize=(10, 6))
sns.barplot(data=month_year_volume_analysis, x='year_month', y='volume', color='c')
plt.title('Monthly Volume Analysis')
plt.xlabel('Year-Month')
plt.ylabel('Total Volume')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
custom_palette = {
    2019: 'yellow',
    2024: 'red',
    2022: 'green',
    2023: 'blue'
}

plt.figure(figsize=(12, 6))
sns.barplot(data=month_year_volume_analysis, x='year_month', y='volume', hue='year', palette=custom_palette)
plt.title('Monthly Volume Analysis')
plt.xlabel('Year-Month')
plt.ylabel('Total Volume')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
