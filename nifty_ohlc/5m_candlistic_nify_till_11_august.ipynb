```python
import pandas as pd
```

```python
df = pd.read_csv('nifty_5m_august_2024.csv')
df.index = pd.to_datetime(df.index)
```

```python
import mplfinance as mpf
```

```python
filtered_df = df[df['hour'].isin([9,10,11])]
```

```python
filtered_df
```

```python
august_dfs = [group for _, group in filtered_df[['year','month','day','open','high','low','close']].groupby('day')]
```

```python
len(august_dfs)
```

```python
import os
```

```python
for august_df in august_dfs:
# Plot candlestick chart
    directory_path = f"{os.getcwd()}/{str(df['year'].iloc[0])}_{str(df['month'].iloc[0])}/5m_candlistic"
    if not os.path.exists(directory_path):
        os.makedirs(directory_path)
    mpf.plot(august_df, type='candle', style='charles', title='Candlestick Chart', ylabel='Price', savefig=os.path.join(directory_path,str(august_df['day'].iloc[0])))
```
