```python
import pandas as pd
```

```python
df_5 = pd.read_csv('nifty_5m_september_2024.csv')
df_5.index = pd.to_datetime(df_5.index)
```

```python
import mplfinance as mpf
```

```python
df_5
# Taking a first three rows of each day.
df_5 = df_5.groupby(df_5['day']).head(1)
df_5
```

```python
import os
```

```python
for index, row in df_5.iterrows():
# Plot candlestick chart
    directory_path = f"{os.getcwd()}/{str(row['year'])}_{str(row['month'])}/ohh_analysis"
    if not os.path.exists(directory_path):
        os.makedirs(directory_path)
    
    # Create a DataFrame with OHLC data for a single day
    ohlc_data = pd.DataFrame({
        'Open': [row['open']],
        'High': [row['high']],
        'Low': [row['low']],
        'Close': [row['close']],
        'Volume': [row['volume']]  # Optional, only if volume is available
    }, index=[pd.Timestamp(f"{row['year']}-{row['month']}-{row['day']}")])
    

    mpf.plot(ohlc_data, type='candle', style='charles', title='Candlestick Chart', ylabel='Price', savefig=os.path.join(directory_path,str(row['day'])))
```
