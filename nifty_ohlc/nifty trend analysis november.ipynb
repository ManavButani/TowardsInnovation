```python
import pandas as pd
```

```python
df = pd.read_csv('nifty_1h_2023.csv')
```

```python
nov = df[df['month']==11].sort_values(by='day', ascending=True)
```

```python
nov_dfs = [group for _, group in nov[['day_name','hour','day','difference', 'trend']].groupby('day')]
total_days = len(nov_dfs)

most_repeating_nov, least_repeating_nov = [], []
for nov_df in nov_dfs:
    nov_up_count = 0
    nov_down_count = 0
    nov_up_volume = 0
    nov_down_volume = 0
    nov_up_hours = []
    sorted_hours = []
    nov_sorted_up_hours = []
    nov_down_hours = []
    nov_sorted_down_hours = []

    print("~~~~~~~~~~~~~~~~",nov_df['day'].iloc[0], nov_df['day_name'].iloc[0]," November ~~~~~~~~~~~~~~~~~~")
    nov_hour_volume = {hour: volume for hour, volume in zip(nov_df['hour'], nov_df['difference'])}
    nov_hour_trend = {hour: trend for hour, trend in zip(nov_df['hour'], nov_df['trend'])}

    print(nov_hour_trend)
    print(nov_hour_volume)
    print("-----------------")


    for hour, trend in nov_hour_trend.items():
        if trend == "UP":
            nov_up_count+=1
            nov_up_volume+=nov_hour_volume.get(hour)
            nov_up_hours.append(hour)
        else:
            nov_down_count+=1
            nov_down_volume+=nov_hour_volume.get(hour)
            nov_down_hours.append(hour)

    for hour in dict(sorted(nov_hour_volume.items(), key=lambda item: item[1])).keys():
        sorted_hours.append(hour)
        if nov_hour_trend.get(hour) == "UP":
            nov_sorted_up_hours.append(hour)
        else:
            nov_sorted_down_hours.append(hour)

    print("UP:::", nov_up_count, nov_up_volume)
    print("DOWN:::", nov_down_count, nov_down_volume)
    print("-----------------")

    print("UP Hours:::", nov_up_hours)
    print("DOWN Hours:::", nov_down_hours)
    print("-----------------")

    print("SORTED UP Hours:::", nov_sorted_up_hours)
    print("SORTED DOWN Hours:::", nov_sorted_down_hours)
    print("-----------------")
    print("SORTED HOURS ASCENDING:::", sorted_hours)
    print("SORTED HOURS DESCENDING:::", list(reversed(sorted_hours)))
```
