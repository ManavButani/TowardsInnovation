```python
import pandas as pd
```

```python
df_30 = pd.read_csv('nifty_30m_august_2024.csv')
df_30.index = pd.to_datetime(df_30.index)
df_5 = pd.read_csv('nifty_5m_august_2024.csv')
df_5.index = pd.to_datetime(df_5.index)
```

```python
import mplfinance as mpf
```

```python
m30_dfs = [group for _, group in df_30[['year','month','day','open','high','low','close']].groupby('day')]
```

```python
len(m30_dfs)
```

```python
df_5
# Taking a first three rows of each day.
df_5 = df_5.groupby(df_5['day']).head(3)
df_5
```

```python
# Add a 'source' column to distinguish between pd1 and pd2
df_5['source'] = 'A'
df_30['source'] = 'B'

# Concatenate both DataFrames
merged_df = pd.concat([df_5, df_30])

# Sort by 'date' and then by 'source' (pd1 rows come before pd2 rows)
merged_df = merged_df.sort_values(by=['day', 'source'], ascending=[True, True])

merged_df
```

```python
import os
```

```python
# Create a list to store the grouped results
august_dfs = []

# Get the unique dates in the DataFrame
unique_days = merged_df['day'].unique()

# Iterate over the dates, excluding the last one
for i in range(len(unique_days) - 1):
    current_day_data = merged_df[merged_df['day'] == unique_days[i]]
    
    # Get the next day's data (first 3 rows)
    next_day_data = merged_df[merged_df['day'] == unique_days[i + 1]].head(3)
    
    # Concatenate current day's data with next day's data
    combined_data = pd.concat([current_day_data, next_day_data])
    
    # Append to the grouped list
    august_dfs.append(combined_data)
```

```python
for august_df in august_dfs:
# Plot candlestick chart
    directory_path = f"{os.getcwd()}/{str(merged_df['year'].iloc[0])}_{str(merged_df['month'].iloc[0])}/up_down_analysis"
    if not os.path.exists(directory_path):
        os.makedirs(directory_path)
    mpf.plot(august_df, type='candle', style='charles', title='Candlestick Chart', ylabel='Price', savefig=os.path.join(directory_path,str(august_df['day'].iloc[0])))
```
