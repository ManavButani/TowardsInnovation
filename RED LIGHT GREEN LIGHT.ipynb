```python
import pandas as pd
```

```python
df = pd.read_csv('NIFTY 50 - first 45 minutes data 2015 to Aug 2024.csv')
df.index = pd.to_datetime(df['date'])
```

```python
df
```

```python
df = df[(df['year'] == 2024) & ~((df['hour'] == 9) & (df['minute'] == 15))]
```

```python
df
```

```python
df['next_trend'] = df['trend'].shift(-1)
df['UPPER_SL'] = df['high'].shift(-1).combine(df['high'], max)
df['LOWER_SL'] = df['low'].shift(-1).combine(df['low'], min)

# Filter rows where the current trend is not equal to the next trend (reciprocal trends)
# reciprocal_pairs = df[df['trend'] != df['next_trend']].iloc[:-1]  # Remove the last row (since it has no "next" row)

# Display the reciprocal trends (current and next rows)
# print(reciprocal_pairs[['date', 'trend', 'next_trend']])
```

```python
final_output = []
for _, group in df.groupby(by=['year','month','day']):
    sub_output = {}
    first_reciprocal_pair_index = group[group['trend'] != group['next_trend']].index[0]
    filtered_df = group.loc[first_reciprocal_pair_index:]
    high = filtered_df.iloc[0]['UPPER_SL']
    low = filtered_df.iloc[0]['LOWER_SL']
    sub_output['year'] = filtered_df.iloc[0]['year']
    sub_output['month'] = filtered_df.iloc[0]['month']
    sub_output['day'] = filtered_df.iloc[0]['day']
    sub_output['hour'] = filtered_df.iloc[0]['hour']
    sub_output['minute'] = filtered_df.iloc[0]['minute']
    SL = None
    BUY_TYPE = None
    executed_after_minutes = 0
    sl_after_minutes = 0
    for i in range(1, len(filtered_df)):
        if SL is None:
            close = filtered_df.iloc[i].close
            open = filtered_df.iloc[i].open
            executed_after_minutes += 1
            if close<=low and open>=high:
                sub_output['is-engullfed'] = True
            else:
                if close<=low:
                    SL = low
                    BUY_TYPE = 'PUT'
                elif open>=high:
                    SL = high
                    BUY_TYPE = 'CALL'
                sub_output['sl'] = SL
                sub_output['option-buy'] = BUY_TYPE
            continue

        sl_after_minutes += 1
        if BUY_TYPE == "PUT":
            sub_high = filtered_df.iloc[i].close
            if sub_high >= SL:
                sub_output['status'] = 'LOSE'
                sub_output['sl-hitted'] = sub_high
                sub_output['tred-running-time'] = sl_after_minutes
                sub_output['tred-run-after'] = executed_after_minutes
                break
        elif BUY_TYPE == "CALL":
            sub_low = filtered_df.iloc[i].close
            if sub_low <= SL:
                sub_output['status'] = 'LOSE'
                sub_output['sl-hitted'] = sub_low
                sub_output['tred-running-time'] = sl_after_minutes
                sub_output['tred-run-after'] = executed_after_minutes
                break
    if SL:
        final_output.append(sub_output)
```

```python
o = pd.DataFrame(final_output)
```
